```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps[mvps["Player"].str.contains("Ä")]
```

```python
mvps["Player"] = mvps["Player"].str.replace("Ä", "ć", regex=False)
```

```python
mvps["Player"] = mvps["Player"].str.replace("Ä", "č", regex=False)
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head()
```

```python
players["Player"].head(30)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.groupby(["Player", "Year"])
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(10)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
players.head(5)
```

```python
combined[pd.isnull(combined["Pts Won"])]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined[pd.isnull(combined["Pts Won"])]
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams["Team"] = teams["Team"].str.replace("\xa0", " ", regex=False)
```

```python
teams.tail(5)
```

```python
del teams["Unnamed: 0"]
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined["Team"].unique()
```

```python
combined = combined[~pd.isnull(combined["Pos"])]
```

```python
combined[pd.isnull(combined["Team"])]
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats.dtypes
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```
